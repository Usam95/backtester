```python
def filtered_df(df):
    filt_df = df.copy()
    cols = ["freq", "outperf_net", "EMA_S", "EMA_L"]
    filt_df = df.loc[:, cols]
    filt_df =  filt_df.sort_values('outperf_net', ascending=False)
    return filt_df
```
